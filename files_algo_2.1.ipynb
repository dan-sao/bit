```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

from glob import glob
import os
```

```python
PATH_TO_DATA = 'C:\\Users\\VyaznikovKS\\Desktop\\_уч\\CRYPTO\\logs'
```

```python
%%time
log = pd.DataFrame(columns=range(70))

for name in glob(PATH_TO_DATA + '\\*.log'):
    
    with open(name, 'r', encoding='utf-8') as file :
        filedata = file.read()
        filedata = filedata.replace(', ', ',')
        filedata = filedata.replace('+ ', '+')
        filedata = filedata.replace('- ', '-')
        filedata = filedata.replace(' = ', ' ')
        filedata = filedata.replace('%', '')
    with open(name, 'w', encoding='utf-8') as file:
        file.write(filedata)  
    
    log_add = pd.read_csv(name,
                          header = None,
                          sep=' ',
                          #skipfooter=400000,
                          names=range(70),
                          skipinitialspace=True,
                          error_bad_lines=False)
    log = pd.concat([log, log_add])

#log.head()
```

```python
%%time
log_1 = log[log[4].isin(['EMAFilter:', 'done,total:','***'])]
log_10 = log[log[4].isin(['EMAFilter:'])]
log_10
```

```python
log_numbers = log_10.iloc[::,6:].dropna(axis=1, how='all')
#log_numbers
```

```python
k= {}
for i in range(min(log_numbers.columns), max(log_numbers.columns), 2):
    k = set().union(k, np.unique(log_numbers[i].dropna().values).tolist())
k = list(k)
```

```python
log_numbers_short = pd.DataFrame(columns = k)
#log_numbers_short
```

```python
for i in range(log_numbers.shape[0]):
    for j in range(min(log_numbers.columns), max(log_numbers.columns), 2):
        #print(i,j)
        #print(log_numbers.loc[i,j])
        #print(log_numbers.loc[i,j+1])
        log_numbers_short.loc[i,log_numbers.loc[i,j]] = log_numbers.loc[i,j+1]
log_numbers_short = log_numbers_short.dropna(axis=1, how='all')
#log_numbers_short
```

```python
dims = pd.concat([log_numbers_short, log_10.iloc[log_10.index][3]], axis=1)
#dims
```

```python
log_2 = log_1[log_1[4] == 'done,total:'][[3,7]].reset_index(drop=True)
log_3 = log_1.iloc[log_1[log_1[4] == 'done,total:'].index - 1][3].reset_index(drop=True)

log_marge = pd.concat([log_2, log_3], axis=1, ignore_index=True)

log_marge.columns = [1, 2, 3]

pd.merge(left=log_marge, right=dims)
```

```python
logi = pd.merge(dims, log_marge)
logi.rename(columns={3: "iter", 1: "coin", 2: "profit_%"}, inplace = True)
```

```python
logi['profit_%'] = logi['profit_%'].str[1:-1]
```

```python
logi['profit_%'] = logi['profit_%'].astype('float')
```
